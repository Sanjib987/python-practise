```python
# # Priority:
# 1 ()
# 2 **
# 3 / * // %
# 4 + -

# in case of same priority, just go left to right
```

---

- Division and multiplication have the same order of precedence
- Addition and subtraction have the same order of precedence
- In case of encountering same precedence operator, go from left to right

image.png

```python
print(5*2+5) # * get's priority >> (5*2)+5
print((5*2)+5)
print(5*2//10) # * and // have the same priority, so go from left to right>> * get's priority
print((5*2)//10)
10  - 4 * 2 + 5 - 6 // 2 # *
# 10-8+5-6//2 # //
# 10-8+5-3
# 2+5-3
# 7-3
# 4
```

```python
# >, greater than
print(5>2)
print(5>5)
print(5>7)

# <, less than
print(5<2)
print(5<5)
print(5<7)

# <=, less than or equal to
print(5<=2)
print(5<=5)
print(5<=7)

# >=, greater than or equal to
print(5>=2)
print(5>=5)
print(5>=7)

# ==, exactly equal to
print(2==5)
print(2==2)
print(2==2.0)
print(2=='2')

# !=, not exactly equal to
print(2!=5)
print(2!=2)
print(2!=2.0)
print(2!='2')
```
